17. Explain the difference between apply() and map() in Pandas?

Ans 1. **map()**:

**Scope**: Exclusively for Pandas Series (single columns).

**Purpose**: Primarily for element-wise value substitution or mapping. It can
replace values in a Series based on a function, a dictionary, or another Series.

**Behavior**: When given a dictionary or Series, map() replaces values in the calling Series with corresponding values from the mapping object. If a key is not found in the mapping, the corresponding value in the Series becomes NaN. When given a function, it applies the function to each element individually.

2. **apply()**:
**Scope**:
Can be used on both Pandas Series and DataFrames.

**Purpose**:
More versatile for applying functions. It can apply a function along an axis (rows or columns) of a DataFrame, or to each element of a Series.

**Behavior**:
When used on a DataFrame, apply() can operate on entire rows or columns, allowing for more complex transformations or aggregations that involve multiple values. When used on a Series, it applies the function to each element, similar to map() with a function.
